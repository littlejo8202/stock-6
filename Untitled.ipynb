```python
import yfinance as yf
import pandas as pd
import os
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# ----------------------------
# 1️⃣ 저장 폴더 지정
# ----------------------------
output_dir = "./etf_daily_returns"
os.makedirs(output_dir, exist_ok=True)

# ----------------------------
# 2️⃣ 테마별 ETF 목록 정의
# ----------------------------

themes = {
    # 전통적 테마
    "Energy": "XLE",
    "Financials": "XLF",
    "Utilities": "XLU",
    "Consumer Staples": "XLP",
    "Industrials": "XLI",
    "Materials": "XLB",
    "Healthcare": "XLV",
    "Telecom": "IYZ",
    "Real Estate": "XLRE",
    "Automotive": "CARZ",

    # 혁신적 테마
    "AI & Robotics": "BOTZ",
    "Clean Energy": "ICLN",
    "Nuclear Energy": "URA",
    "Biotech": "IBB",
    "EV & Battery": "LIT",
    "Semiconductors": "SOXX",
    "Cybersecurity": "CIBR",
    "Space Industry": "UFO",
    "Carbon Neutral": "KRBN",
    "Fintech": "FINX",
}

# ----------------------------
# 3️⃣ 야후 파이낸스 데이터 다운로드
# ----------------------------
start_date = "2015-01-01"
end_date = "2025-01-01"

for theme, ticker in themes.items():
    print(f"📥 {theme} ({ticker}) 데이터 다운로드 중...")

    # auto_adjust=False → 'Adj Close' 포함, True → 'Close'만 포함
    df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=True)

    if df.empty:
        print(f"⚠️ {ticker} 데이터가 없습니다. 건너뜁니다.")
        continue

    # ----------------------------
    # 4️⃣ 일간 수익률 계산 ('Close' 기준)
    # ----------------------------
    df["Daily Return (%)"] = df["Close"].pct_change() * 100

    # ----------------------------
    # 5️⃣ 파일 저장
    # ----------------------------
    file_name = f"{output_dir}/{ticker}_daily_returns.csv"
    df.to_csv(file_name, index=True)
    print(f"✅ 저장 완료 → {file_name}")

print("\n🎯 모든 ETF의 일간 수익률 데이터 저장 완료!")

```
